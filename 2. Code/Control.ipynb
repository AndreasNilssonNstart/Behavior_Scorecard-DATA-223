```python
def fn_amortize(base, amount, m):
    i = 1
    amort = 0.0
    while i <= m:
        amort += (amount - amort) * (1.0 / base)
        i += 1

        print(amort)

    return round(amort, 2)

# Example usage:
base = 36
amount = 1000.00  # Example amount
m = 12  # Example period

amortized_value = fn_amortize(base, amount, m)
print(amortized_value)
```
